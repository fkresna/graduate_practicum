```python
import csv, json, ast
from pymongo import MongoClient
```

```python
client = MongoClient()
db = client.ecommerceDB

#Delete customer table if exist, This deletion prevent data duplication
db.customer.delete_many({})

name = []
province = []
region = []
segment = []
email = []
#start import data from csv
with open("../../rawDataSet/superstoreSalesData/order.csv") as f:
	reader = csv.reader(f)
	count = 0
	unique = 0
	for row in reader:
        #First row is column name
		if count != 0:
            #Make sure the customer has not on the list yet
			if row[11] not in name:
				name.append(row[11])
				province.append(row[12]);
				region.append(row[13]);
				#Put [name]@gmail.com on email_address field, because originally there are no email_address 
				#in origial datase
				email.append(row[11].replace(" ", "").lower()+"@gmail.com");
				#print name[unique]
				unique = unique + 1
				segment.append(row[14])
		count = count + 1
print unique
k = unique-1
jsonString = {'customer_id':k,'customer_group_id':segment[k],'name':name[k],'email_address':email[k],'province':province[k],'region':region[k],'country':'CA'}
print jsonString
```

```python
#Insert JSON data into customer table
for j in range(0,len(name)):
	jsonString = {'customer_id':j,'customer_segment':segment[j],'name':name[j],'email_address':email[j],'province':province[j],'region':region[j],'country':'CA'}
	result = db.customer.insert_one(jsonString)
	print jsonString
	print result
	print j
```
